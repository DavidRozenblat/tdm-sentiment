### Setup

```python
# import libraries 
from pathlib import Path
import sys
import pandas as pd
from copy import copy
import ast
import matplotlib.pyplot as plt
import seaborn as sns
import hashlib
import numpy as np


# import project libraries
project_path = Path('/home/ec2-user/SageMaker/david/tdm-sentiment/')
sys.path.append(str(project_path / 'code'))
from tdm_parser.tdm_parser import TdmXmlParser #TODO

# handle path
data_path = project_path / 'data/'
corpus_name = 'representative_sample' #TODO  # 'USATodayDavid' ChicagoTribune' 'Newyork20042023' TheWashingtonPostDavid, LosAngelesTimesDavid, StartupSentiment
folder_path = data_path / 'results/' / corpus_name

```

```python
def get_all_csv_files(folder_path):
    # Use Path to find all CSV files in the folder
    csv_files = list(Path(folder_path).glob('*.csv'))
    return csv_files



def concat_csv_files(folder_path):
    # Get a list of all CSV files in the folder
    csv_files = get_all_csv_files(folder_path)
    # Initialize an empty list to store DataFrames
    df_list = []
    # Loop through the list of CSV files
    for file in csv_files:
        # Read each CSV file into a DataFrame
        df = pd.read_csv(file)
        # Append the DataFrame to the list
        df_list.append(df)
    # Concatenate all DataFrames in the list into a single DataFrame
    combined_df = pd.concat(df_list, ignore_index=True)
    return combined_df


def md5_hash_text(text):
    # If 'text' is a list, join its items into a single string.
    if isinstance(text, list):
        # Convert each item to string (in case they're not),
        # then join them with a space (or any delimiter you want).
        text = " ".join(map(str, text))
    
    return hashlib.md5(text.encode('utf-8')).hexdigest()


def get_text_string(text):
    text = " ".join(map(str, text))
    return text

def safe_literal_eval(input_str):
    """Safely evaluate a string containing a Python literal or container display."""
    try:
        return ast.literal_eval(input_str)
    except:
        return input_str
    
def article_size(word_count):
    """
    Classify articles as small (<300 words), medium (300–1000 words), or large (>1000 words).
    """
    if word_count < 300:
        return 'small'
    elif word_count > 1000:
        return 'large'
    return 'medium'
```

```python
# Combine CSV files from all files
combined_df = concat_csv_files(folder_path)

# filter and clean df 
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df = combined_df[combined_df['Date'].dt.year < 2024]
combined_df = combined_df[combined_df['Date'].dt.year > 2003]

combined_df['Publisher'] = combined_df['Publisher'].replace( # handle typos and change of terms over time 
    {'international herald tribune': 'new york times',
     'new york times the': 'new york times',
     'international new york times': 'new york times',
     'los angeles times (pre-1997 fulltext)': 'los angeles times',
     'washington post (pre-1997 fulltext) the': 'washington post the' }
)
combined_df = combined_df[combined_df['Publisher'] != 'new york times español']
combined_df['Text'] = combined_df['Text'].apply(safe_literal_eval)
combined_df = combined_df[combined_df[['Text', 'Page', 'Date', 'WordCount']].notna().all(axis=1)]
combined_df["text_hash"] = combined_df["Text"].apply(md5_hash_text)
combined_df.drop_duplicates(subset=['text_hash'], inplace=True)
combined_df = combined_df.drop(columns=['text_hash'])

#combined_df.shape
combined_df.head(1)
```

### Data Descriptive Statistics 

```python
# plot number of articles over time 
df = combined_df.copy()
df['Year'] = df['Date'].dt.year
df = df.groupby(['Year']).size().reset_index(name="Count")

plt.figure(figsize=(20,10))
sns.barplot(x='Year', y='Count', data=df)
plt.title('Trend of article counts over time')
plt.show()
```

```python
# plot number of articles over time 
df = combined_df.copy()
df['Year'] = df['Date'].dt.year
df = df.groupby(['Year', 'Publisher']).size().reset_index(name="Count")

plt.figure(figsize=(20,10))
sns.barplot(x='Year', y='Count', hue='Publisher', data=df)
plt.title('Trend of article counts over time by publisher')
plt.show()
```

```python
# plot number of articles over time by article size
df = combined_df.copy()
df['article_size'] = df['WordCount'].apply(article_size)
df['Year'] = pd.to_datetime(df['Date']).dt.year
df = df.groupby(['Year', 'article_size']).size().reset_index(name="Count")

plt.figure(figsize=(20,10))
sns.barplot(
    x='Year',
    y='Count',
    hue='article_size',
    data=df,
    hue_order=['large', 'medium', 'small'],  # Specify the desired order
    palette={'small': 'green', 'medium': 'orange', 'large': 'blue'}
)
plt.title('Trend of article counts over time by article size')
plt.show()
#df
```

### Model descriptive Statistics

```python
# plot number of articles over time by model recognition
df = combined_df.copy()
df['Year'] = pd.to_datetime(df['Date']).dt.year
df = df.groupby(['Year', 'is_economic']).size().reset_index(name="Count")

plt.figure(figsize=(20,10))
sns.barplot(x='Year', y='Count', hue='is_economic', data=df)
plt.title('Trend of article counts over time by model recognition')
plt.show()
```

```python
# plot ratio of articles over time by model recognition
df = combined_df.copy()
df['Year'] = pd.to_datetime(df['Date']).dt.year
df = df.groupby(['Year', 'is_economic'])['GOID'].size().reset_index(name="Count")
df = df.pivot_table(index='Year', columns='is_economic', values='Count')
df['ratio'] = df[1.0]/(df[0.0]+ df[1.0])

plt.figure(figsize=(20,10))
sns.barplot(x='Year', y='ratio', data=df)
plt.title('ratio of economic articles over time')
plt.show()
```
